```python
data=data[(data['age']<=75)&(data['age']>=17)]
```

```python
plt.boxplot(data['age'])
plt.show()
```

```python
#encoding-> converting Categorical value into numerical value
#label encoding
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
data['workclass']=encoder.fit_transform(data['workclass'])
data['marital-status']=encoder.fit_transform(data['marital-status'])
data['occupation']=encoder.fit_transform(data['occupation'])
data['relationship']=encoder.fit_transform(data['relationship'])
data['race']=encoder.fit_transform(data['race'])
data['gender']=encoder.fit_transform(data['gender'])
data['native-country']=encoder.fit_transform(data['native-country'])
data
```

```python
x=data.drop(columns=['income']) #input 
y=data['income']  #output variable
x
```

```python
y
```

```python
#scaler technique
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler() #data will be converted from 0-1
x=scaler.fit_transform(x)
x
```

```python
#spliting data into training and testing
from sklearn.model_selection import train_test_split
xtrain, xtest,ytrain,ytest=train_test_split(x,y, test_size=0.2, random_state=23,stratify=y)
```

```python
xtrain
```

```python
#machine learning algorithm
from sklearn.neighbors import KNeighborsClassifier  # supervised classifier

knn = KNeighborsClassifier()
knn.fit(xtrain, ytrain)  # training the model
predict = knn.predict(xtest)  # predicting on test data
predict
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(ytest,predict)
```

```python
#logicstic regression
from sklearn.linear_model import LogisticRegression  # supervised classifier

lr = LogisticRegression()
lr.fit(xtrain, ytrain)       # training the model
predict1 = lr.predict(xtest) # predicting on test data
predict1
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(ytest,predict1)
```

```python
from sklearn.neural_network import MLPClassifier
clf=MLPClassifier(solver='adam',hidden_layer_sizes=(5,3), random_state=2, max_iter=2000) #5-neurons and 3- hidden layers
clf.fit(xtrain, ytrain)
predict2=clf.predict(xtest)
predict2
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(ytest,predict2)
```
